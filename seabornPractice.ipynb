```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
```

Line Plot

```python
a = [1, 2, 3, 4, 5, 6, 7]
b = [2, 5, 7, 6, 2, 1, 4]
```

```python
# using only matplotlib
plt.plot(a, b)
plt.show()    #  show() function is used to display the plot created by Matplotlib
```

```python
# using seaborn
df = pd.DataFrame({'var1':a, 'var2':b})
sns.lineplot(x='var1', y='var2', data=df)
```

```python
# getting data from external data set i.e from github, file name is titanic.csv
data_set = sns.load_dataset('titanic')
data_set.head()
```

```python
sns.lineplot(x='pclass', y='fare', data=data_set)
```

```python
# hue parameter in seaborn allows for an additional layer of categorical differentiation in your plots,
# making them more informative and easier to understand.
sns.lineplot(x='alive', y='age', data=data_set, hue='sex')
```

```python
plt.figure(figsize=(12, 6))  # set the size of the figure, it should be declared before creting the plot

# parameter, Seaborn will apply different line styles or marker styles to the plot elements based on the unique values of the specified variable.
sns.lineplot(x='survived', y='age', data=data_set, hue='sex', style='embark_town')
```

```python
sns.lineplot(x='survived', y='age', data=data_set, hue='sex', style='embark_town', legend=False)
```

Bar Plot

```python
data_set.columns
```

```python
sns.barplot(x='class', y='fare', data=data_set, hue='sex')
```

Count Plot

A count plot is used to show the count of observations in each categorical bin using bars.
A bar plot is used to show the relationship between a categorical variable and a numeric variable using bars.

```python
sns.countplot(x='sex', data=data_set)
```

```python
sns.countplot(x='sex', data=data_set, hue='survived')
```
